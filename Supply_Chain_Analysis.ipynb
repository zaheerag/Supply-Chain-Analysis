```python
# plan stage
## import packages

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go

pio.templates.default = "plotly_white"
```

```python
## read the data

df = pd.read_csv('supply_chain_data.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
## data cleaning
### look for missing values.

df.isnull().sum()
```

```python
### look for and duplicates and drop them

df.drop_duplicates(keep = 'first')
```

```python
### detect any outliers

fig_outlier = px.box(df, y="Shipping costs", title="Detecting Shipping Cost Outliers")
fig_outlier.show()
```

```python
# calculate how much revenue we get per Rand spent on shipping
df['Cost_Efficiency'] = df['Revenue generated'] / df['Shipping costs']

# filter out rows where shipping cost is 0, to avoid dividing by 0
df = df[df['Shipping costs'] > 0]
```

```python
df.head()
```

```python
## deal with "at risk" orders

at_risk_df = df[(df['Lead time'] > 15) | (df['Stock levels'] < 20)].copy()
at_risk_df = at_risk_df.sort_values(by='Stock levels', ascending=True)
table_data = at_risk_df[['SKU', 'Product type', 'Stock levels', 'Lead time', 'Location']]
```

```python
### create "at risk" table

import plotly.graph_objects as go

fig_table = go.Figure(data=[go.Table(
    header=dict(values=list(table_data.columns),
                fill_color='#008de7',
                font=dict(color='white', size=12),
                align='left'),
    cells=dict(values=[table_data[col] for col in table_data.columns],
               fill_color='#f5f5f5',
               align='left'))
])

fig_table.update_layout(title="Priority Action Items (Low Stock / High Lead Time)")
fig_table.show()
```

```python
## revenue vs cost

df['Logistics_Ratio'] = (df['Shipping costs'] / df['Revenue generated']) * 100

fig_efficiency = px.scatter(df,
                 x='Revenue generated',
                 y='Shipping costs',
                 color='Product type',
                 size='Stock levels',
                 hover_data=['SKU'],
                 title='Shipping Cost vs. Revenue Efficiency',
                 labels={'Revenue generated': 'Revenue ', 'Shipping costs': 'Shipping Cost '},
                 trendline="ols")

fig_efficiency.show()
```

```python
def generate_daily_report(df):

    total_revenue = df['Revenue generated'].sum()
    avg_lead_time = df['Lead time'].mean()
    at_risk_count = len(df[df['Stock levels'] < 20])

    print(f"--- Daily Report ---")
    print(f"Total Revenue: R{total_revenue:,.2f}")
    print(f"Avg Lead Time: {avg_lead_time:.2f} Days")
    print(f"Critical Stock Items: {at_risk_count}")

    return 'Report Generated Successfully'

generate_daily_report(df)
```
